```python
from dateutil.parser import parse
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
plt.rcParams.update({'figure.figsize': (10, 7), 'figure.dpi': 120})
sns.set_style('whitegrid')
df = pd.read_csv(
'https://raw.githubusercontent.com/selva86/datasets/master/a10.csv',
parse_dates=['date'],
index_col='date'
)
print("First 5 rows of dataset:\n")
print(df.head())
print("\nDataset Information:")
print(df.info())
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16, 5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
plot_df(
df,
x=df.index,
y=df.value,
title='Monthly Anti-Diabetic Drug Sales in Australia (1992–2008)'
)
df_market = pd.read_csv(
'https://raw.githubusercontent.com/selva86/datasets/master/MarketArrivals.csv'
)
df_mumbai = df_market.loc[df_market.market == 'MUMBAI', :]
print("\nFirst 5 rows of Mumbai Market Data:\n")
print(df_mumbai.head())
print("\nNumber of records for Mumbai market:", len(df_mumbai))
```
